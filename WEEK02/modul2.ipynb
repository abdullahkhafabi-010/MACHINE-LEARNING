# Tugas

Deskripsi Tugas

Pada tugas pratikum ini Anda akan menggunakan data "Wisconsin Breast Cancer". Data tersebut terdiri dari 569 data yang digunakan untuk mendiagnonis jenis kanker Malignant (M) dan Benign (B). Tugas Anda adalah,

1. Pisahkan antara variabel yang dapat digunakan dan variabel yang tidak dapat digunakan.
2. Lakukan proses encoding pada kolom "diagnosis".
3. Lakukan proses standarisasi pada semua kolom yang memiliki nilai numerik.
4. Lakukan proses stratified split data untuk membuat data latih dan data uji dengan rasio 80:20.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot
from sklearn.preprocessing import LabelEncoder, Normalizer
from sklearn.model_selection import train_test_split

dpath = 'wbc.csv'
df = pd.read_csv(dpath)
df.head()
```

```python
df.info()
```

```python
df.isnull().sum()
```

Pisahkan antara variabel yang dapat digunakan dan variabel yang tidak dapat digunakan

```python
# Meng-copy data dari df
df2 = df.copy()
df2.head()

# Mengambil data yang digunakan & data yang tidak digunakan
# Menghapus data yang tidak digunakan
unused_columns = ['id']
unnamed_columns = ['Unnamed: 32']
df2.drop(unused_columns, axis=1, inplace=True)
df2.drop(unnamed_columns, axis=1, inplace=True)
df2.head()
```

Lakukan proses encoding pada kolom "diagnosis"

```python
# Import LabelEncoder
from sklearn.calibration import LabelEncoder

# Encoding data menggunakan LabelEncoder
le = LabelEncoder()
df2['diagnosis'] = le.fit_transform(df2['diagnosis'])
df2.head()
```

Lakukan proses standarisasi pada semua kolom yang memiliki nilai numerik

```python
from sklearn.discriminant_analysis import StandardScaler

std = StandardScaler()
df2.iloc[:, 1:] = std.fit_transform(df2.iloc[:, 1:]) # Mengubah data menjadi data standar untuk semua kolom kecuali kolom pertama yaitu diagnosis
df2.head()
```

Lakukan proses stratified split data untuk membuat data latih dan data uji dengan rasio 80:20

```python
from sklearn.model_selection import train_test_split

# Memisahkan data menjadi data training dan data testing
df2_train, df2_test = train_test_split(df2, test_size=0.2, random_state=0, stratify=df2['diagnosis'])

# Cek ukuran data training dan data testing
print(f'Jumlah label data asli:\n{df2.diagnosis.value_counts()}\n')
print(f'Jumlah label data train:\n{df2_train.diagnosis.value_counts()}\n')
print(f'Jumlah label data test:\n{df2_test.diagnosis.value_counts()}')
```
